```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r'C:\DATA\mtcars.csv')
df
```

- Q1

```python
plt.scatter(df["hp"],df["mpg"],c=df["am"])
plt.show()
```

- Q2

```python
x=[df.mpg[df.cyl==i] for i in np.unique(df.cyl)]
plt.boxplot(x)
plt.title('Boxplot')
plt.xticks([1,2,3],['2','4','6'])
plt.xlabel('cyl',c='r')
plt.ylabel('mpg',c='r')
plt.show()
```

- Q3

```python
plt.hist(df["wt"],color="maroon",edgecolor='k',bins=10,alpha=0.7)
plt.xlabel("WEIGTH")
plt.title("Histogram")
plt.axvline(df["wt"].mean(),color='k',linestyle='dashed')

```

- Q4


```python
#stacked bar chart using pandas
df1=df.groupby('cyl')['hp','wt'].mean()
df1.plot(kind='bar',stacked=True,edgecolor='k')

```

- Q6


```python
df4=df[['mpg','hp','wt','cyl','am']].copy()
pd.plotting.scatter_matrix(df4,figsize=(5,5),grid=True,marker='cyl',c=df['am'])
plt.show()
```

- Q7


```python
corr_matrix=df.corr(method='pearson')
#plt.imshow(corr_matrix,cmap='autumn',interpolation='nearest')
#plt.show()
fig,ax=plt.subplots(figsize=(4,4))    
ax.imshow(corr_matrix,cmap='autumn',interpolation='nearest')
ax.set_xticklabels(corr_matrix.columns,fontsize=10)
ax.set_yticklabels(corr_matrix.columns,fontsize=10)
plt.show()
```

```python
y1=pd.DataFrame([df.mpg[df.am==1] for i in [1]])
y2=pd.DataFrame([df.mpg[df.am==0] for i in [0]])
y=pd.DataFrame([y1.mean(axis=1),y2.mean(axis=1)])
plt.bar(y.index,y.mpg,edgecolor="k")
plt.title("barplot")
plt.ylabel("mean_of_mpg(automatic vs manual)")
plt.xlabel("am")
plt.show()
```
