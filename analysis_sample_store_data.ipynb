# Exploratory Data Analysis(Sample_store_data)


### importing python libraries 

```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
```

### Read sample store data 

```python
df=pd.read_csv(r'C:\DATA\SampleSuperstore.csv')
data=df.drop(["Postal_Code"],axis=1)
data.head()
```

### Basic information about data set

```python
data.shape
```

## Descriptive Statistics

```python
data.describe()
```

```python
data.groupby(["Ship_Mode","Segment","City","Region","Category"]).describe()

```

```python
data.corr()
```

## Bar plot

```python
plt.bar("Sub_Category","Category",data=data)
plt.xticks(rotation=90)
plt.show()
```

## Pie chart (Sub_Category)

```python
plt.pie(data["Sub_Category"].value_counts(),labels=np.unique(data.Sub_Category))
plt.show()
```


## Count Plot (State)

```python
print(data["State"].value_counts())
plt.figure(figsize=(12,10))
sns.countplot(x=data.State)
plt.xticks(rotation=90)
plt.show()
```

```python
print(data.Sub_Category.value_counts())
sns.countplot(x=data.Sub_Category)
plt.xticks(rotation=90)
plt.show()
```

## Bar plot (Group_wise)

```python
x=data.groupby("Region")["Sales","Profit"].mean()
#x.plot(kind='bar',stacked=True,edgecolor='k')
x.plot(kind='bar',edgecolor='k')

```

## Heat map

```python
data.corr()
sns.heatmap(data.corr(),annot=True)
plt.title("Heat_Map")
plt.show()
```

## Histogram (Profit)

```python
plt.figure(figsize=(8,6))
sns.histplot(data = data, x = "Profit", kde = True,stat="probability",hue="Region")
plt.axvline(np.mean(data.Sales),color="k",linestyle="dashed")

```

## Box plot (Category)

```python
plt.figure(figsize=(8,6))
plt.boxplot([data.Sales[data.Category==i] for i in np.unique(data.Category)],notch=True)
plt.show()
```
